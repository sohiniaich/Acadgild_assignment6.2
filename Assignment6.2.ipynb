```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url= 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv' 
titanic = pd.read_csv(url) 

```

```python
plt.pie(titanic.sex.groupby(titanic.sex).count(),labels=titanic.sex.groupby(titanic.sex).count().index,
        shadow=False, radius=1.25,autopct='%.2f', startangle=90,colors=["Magenta","Blue"])
plt.title('Gender Percentage')
plt.show()
```

```python
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age Vs Fare by Gender")
titanic_male=titanic[titanic['sex']=="male"]
titanic_female=titanic[titanic['sex']=="female"]
plt.scatter(titanic_male.age,titanic_male.fare,c='Blue',marker='o')
plt.scatter(titanic_female.age,titanic_female.fare,c='Pink',marker='o')
plt.legend(titanic.sex)
plt.show()
```
